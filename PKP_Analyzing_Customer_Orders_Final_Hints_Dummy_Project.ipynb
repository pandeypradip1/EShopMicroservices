# Course End Project - Analyzing Customer Orders Using Python

## Overview
Analyze customer orders using Python data structures to classify products, identifying customer purchasing patterns, and generate business insights.

Tasks
1. **Store Customer Orders**
2. **Classify products by category**
3. **Analyze customer orders**
4. **Generate business insights**
5. **Organize and display data**

```python
# --- Step 1: Store customer orders

# -- List of customer names
customers = ["Allan", "Boby", "Champ", "Donny", "Ewan"]

# -- customer's order details (customer name, product, price, category) as tuples inside a list
orders = [    ("Allan","Laptop",250, "Electronics"),
    ("Allan","Mobile",100, "Electronics"),
    ("Allan","Headphones",200, "Electronics"),
    ("Boby","T-shirts",50, "Clothing"),
    ("Boby","Jeans",100, "Clothing"),
    ("Boby","Air conditioners",200, "Electronics"),
    ("Boby","microwave",100, "Home Essentials"),
    ("Champ", "Vacuum Cleaner", 50, "Home Essentials"),
    ("Champ", "Frying Pan", 30, "Home Essentials"),
    ("Champ", "Washing Machines", 220, "Electronics"),
    ("Donny", "Vegetable peeler", 10, "Home Essentials"),
    ("Donny","Charger",10, "Electronics"),
    ("Donny","Laptop",30, "Electronics"),
    ("Donny","Jeans",10, "Clothing"),
    ("Ewan","Jeans",10, "Clothing"),
    ("Ewan","Jacket",20, "Clothing"),
    ("Ewan","Sweater",10, "Clothing"),
    ("Ewan", "Cook ware", 5, "Home Essentials"),
]

# - A dictionary where keys are customer names and values are lists of ordered products

# - An Empty dictionary
customer_orders = {}
# - For loop to append customer name and values of ordered products in list
for name, product, price, category in orders:
  customer_orders.setdefault(name, []).append(product)

print("Customer Order Dictionary:")
print(customer_orders)
```

```python
# --- Step 2. Classify products by category
# --- An empty dictionary
unique_categories = {}

# - • Use a dictionary to map each product to its respective category
for name, product, price, category in orders:
  unique_categories[product] = category

# - • Create a set of unique product categories
unique_prod_categories = set(unique_categories.values())

# - • Display all available product categories
print("\n Unique Product Categories:", unique_prod_categories)
```

```python
# --- 3. Analyze customer orders
# --- • Use a loop to calculate the total amount each customer spends
customer_spending = {}

for name, product,price, category in orders:
  if name not in customer_spending:
    customer_spending[name] = 0
  customer_spending[name] += price

print(customer_spending)

# --- • If the total purchase value is above $100, classify the customer as a high-value buyer
# --- • If it is between $50 and $100, classify the customer as a moderate buyer
# --- • If it is below $50, classify them as a low-value buyer

customer_classification = {}

for name, total in customer_spending.items():
  if total > 100:
    customer_classification[name] = ' $' + str(total) + ' ->' " High Value Buyer"
  elif 50 <= total <= 100:
     customer_classification[name] = ' $' + str(total) + ' ->' "Moderate Value Buyer"
  else:
    customer_classification[name] = ' $' + str(total) + ' ->' "Low Value Buyer"

print("\nCustomer Classification:")
print(customer_classification)
```

```python
# --- Step 4: Generate Business Insights ---
# --- • Calculate the total revenue per product category and store it in a dictionary
# --- • Extract unique products from all orders using a set
# --- • Use a list comprehension to find all customers who purchased electronics
# --- • Identify the top three highest-spending customers using sorting

# --- • Calculate the total revenue per product category and store it in a dictionary
# --- An Empty dictionary
category_revenue = {}

# --- for loop to calculate total revenue per product category
for name, product, price, category in orders:
    if category not in category_revenue:
        category_revenue[category] = 0
    category_revenue[category] += price

print("\nTotal revenue per product category:")
print(category_revenue)

# --- • Extracting unique products from all orders using a set
unique_products = set([product for _, product, _, _ in orders])

print("\nUnique products:")
print(unique_products)

# --- • Use a list comprehension to find all customers who purchased electronics
electronics_customers = list(set([name for name, product, price, category in orders
                                  if category == 'Electronics']))
print("\nCustomers who bought Electronics:")
print(electronics_customers)

# --- • Identify the top 3 highest-spending customers using sorting
# top_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True)[:3]

top_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True)[:3]

print("\nTop 3 highest-spending customers:")
print(top_customers)
```

```python

# --- Step 5: Organize and Display Data ---
# --- • Print a summary of each customer’s total spending and their classification
# --- • Use set operations to find customers who purchased from multiple categories
# --- • Identify common customers who bought both electronics and clothing

# --- • Print a summary of each customer’s total spending and their classification
print("\nCustomer Summary:")
for customer in customers:
    total = customer_spending.get(customer, 0)
    category = customer_classification.get(customer, "No Data")
    print(f"{customer}: ${total} → {category}")

# --- • Use set operations to find customers who purchased from multiple categories
customer_categories = {}

for name, product, price, category in orders:
  if name not in customer_categories:
    customer_categories[name] = set()
  customer_categories[name].add(category)

multi_category_customers = [c for c, cats in customer_categories.items() if len(cats) > 1]

print("\n Customers who purchased from multiple categories:")
print(multi_category_customers)

# --- • Identify common customers who bought both electronics and clothing
electronics_buyers = set()
clothing_buyers = set()

for name, product, price, category in orders:
    if category == "Electronics":
        electronics_buyers.add(name)
    elif category == "Clothing":
        clothing_buyers.add(name)

common_customers = electronics_buyers.intersection(clothing_buyers)

print("\nCustomers who bought Electronics AND Clothing:")
print(common_customers)
```
